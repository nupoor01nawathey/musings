```python
source_df.rename(columns=lambda x: x + '_sf', inplace=True)
target_df.rename(columns=lambda x: x + '_adb', inplace=True)

outer = source_df.merge(
                        right=target_df, 
                        left_on=source_df.columns.to_list(),
                        right_on=target_df.columns.to_list(),
                        how='outer'
                       )
```

```python
print(outer)
```

```python
in_source_not_in_target = outer.loc[outer[target_df.columns.to_list()]
                                        .isnull() .all(axis=1), source_df.columns.to_list()]

print(in_source_not_in_target)
```

```python
in_target_not_in_source = outer.loc[outer[source_df.columns.to_list()]
                                        .isnull() .all(axis=1), target_df.columns.to_list()]

print(in_target_not_in_source)
```
